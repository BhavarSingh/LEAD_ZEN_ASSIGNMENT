```python
import random

# Question Templates
def job_conv_gen(education, job, location):
  # Are you in mumbai and have a computer science degree looking for a developer job ? 
  # Are you a computer science graduate in Mumbai seeking for work as a developer?
  # Do you live in Mumbai and have a degree in computer science?
  # Have been you searching for a developer job in Mumbai and have a computer science degree?
  
  sent_1 = "Are you in " + location + " and have a " + education  + " degree and looking for a "+ job + " job ?"
  sent_2 = "Are you a " + education + " graduate in " + location + " seeking for work as a " + job + " ?"
  sent_3 = "Do you live in " + location + " have a degree in " + education + " and want a " + job+ " job ?"
  sent_4 = "Have been you searching for a " + job +" job in "+ location+" and have a "+education+" degree ? "
  list_of_sent = [sent_1, sent_2, sent_3,sent_4]
  result = random.randint(0,3)
  return list_of_sent[result]

```

```python
# importing required libraries
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
# initating the Parrot model
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
# test sentence for paraphrasing
phrases = ["Do you live in Amritsar have a degree in B.COM and want a Accountant ?"]
```

```python
# check the paraphrased sentences
for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase)
  for para_phrase in para_phrases:
   print(para_phrase)
```
